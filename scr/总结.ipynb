# 机器学习评估指标详解

## AUROC (Area Under the Receiver Operating Characteristic Curve)

**AUROC** 是接收者操作特征曲线下的面积，用于评估**二分类模型**的性能。

- **含义**：衡量模型在不同分类阈值下的真正例率（TPR）和假正例率（FPR）的权衡
- **取值范围**：0 到 1
- **解释**：
    - AUROC = 1：完美分类器
    - AUROC = 0.5：随机分类器
    - AUROC > 0.5：模型优于随机
- **优势**：对类别不平衡不敏感

## AUPRC (Area Under the Precision-Recall Curve)

**AUPRC** 是精确率-召回率曲线下的面积，适合**不平衡数据集**。

- **含义**：衡量精确率（Precision）和召回率（Recall）之间的权衡
- **取值范围**：0 到 1
- **解释**：
    - AUPRC = 1：完美分类器
    - AUPRC 越高：模型越好
- **优势**：对正例较少的场景更敏感

## 相关指标

| 指标 | 定义 | 公式 |
|------|------|------|
| **精确率 (Precision)** | 预测为正的样本中，实际为正的比例 | TP / (TP + FP) |
| **召回率 (Recall)** | 实际为正的样本中，预测正确的比例 | TP / (TP + FN) |
| **F1-Score** | 精确率和召回率的调和平均数 | 2 × (P × R) / (P + R) |
| **准确率 (Accuracy)** | 正确预测的比例 | (TP + TN) / 总数 |

其中：TP=真正例，FP=假正例，TN=真负例，FN=假负例
